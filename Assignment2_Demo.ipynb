```python
## Import Pandas:

import pandas as pd
```

```python
## Import Pyplot:

from matplotlib import pyplot as plt
```

```python
## Read and store the csv file using the path and a variable:

MyData = pd.read_csv("/content/sample_data/students.csv")
```

```python
## Review and check the data:

MyData.head(10)
```

```python
## Create a bar chart for to compare head counts of Undergraduates in 2014:

x = MyData.columns[1:5]   ## Columns
y = MyData.iloc[4, 1:5]  ## Row

plt.bar(x,y)
plt.title("Headcounts of Undergraduates from 2010 to 2019")
plt.xlabel("Years")
plt.ylabel("Headcounts")
plt.show()

```

```python
## Create a line chart to illustrate change of headcounts except Senior over time:

x1 = MyData.iloc[:, 0]
y1 = MyData.iloc[:, 1:4]
```

```python
plt.plot(x1, y1, linestyle = "-", marker = "*")
plt.legend(MyData.columns[1:4])
plt.title("Changes in Headcounts of Undergraduates in 2014")
plt.xlabel("Years")
plt.ylabel("Headcounts")

plt.show()
```

```python
 ## Create a pie chart for headcounts in 2018:

x3 = MyData.columns[1:5]   ## Columns
y3 = MyData.iloc[8, 1:5]  ## Row

plt.pie(y3, labels = x3, autopct = "%0.2f%%")
plt.title("Headcounts of Undergraduates in 2018")
plt.show()
```
