```python
## Import Pandas:

import pandas as pd
```

```python
## Import Pyplot:

from matplotlib import pyplot as plt
```

```python
## Read and store the csv file using the path and a variable:

MyData = pd.read_csv("/content/sample_data/Headcounts.csv")
```

```python
## Review and check the data:

MyData.head(10)
```

```python
## Create a bar chart for to compare head counts among departments in 2014:

x = MyData.columns[1:5]   ## Columns
y = MyData.iloc[2, 1:5]  ## Row

plt.bar(x,y)
plt.title("Headcounts of departments in 2014")
plt.xlabel("Departments")
plt.ylabel("Headcounts")

plt.show()

```

```python
## Create a line chart to illustrate change of headcounts over time:

x1 = MyData.iloc[:, 0]
y1 = MyData.iloc[:, 1:]
```

```python
plt.plot(x1, y1, linestyle = "-", marker = "*")
plt.legend(MyData.columns[1:5])

plt.show()
```

```python
 ## Create a pie chart for headcounts in 2018:

x = MyData.columns[1:5]   ## Columns
y = MyData.iloc[8, 1:5]  ## Row

 plt.pie(y, labels = x, autopct = "%0.2f%%")
 plt.show()
```
