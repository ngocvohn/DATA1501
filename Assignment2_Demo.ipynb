```python
## Import pandas library:
import pandas as pd
```

```python
## Import library for plotting:
from matplotlib import pyplot as plt
```

```python
## Upload and read the dataset:
MyData = pd.read_csv("/content/drive/MyDrive/DATA 1501/students.csv")
```

```python
## Review and verify the dataset:
MyData.head(10)
```

```python
## Plot the bar graph for headcount across departments in 2012
x = MyData.columns[1:5]   ## Columns
y = MyData.iloc[4, 1:5]  ## Row

plt.bar(x,y)
plt.title("Headcounts of departments in 2012")
plt.xlabel("Departments")
plt.ylabel("Headcounts")

plt.show()
```

```python
## Create a line chart to illustrate change of headcounts over time:

x1 = MyData.iloc[:, 0]
y1 = MyData.iloc[:, 1:]

plt.plot(x1, y1, linestyle = "-", marker = "*")
plt.legend(MyData.columns[1:5])

plt.show()
```

```python
## Create a pie chart for headcounts in 2012:

plt.pie(y, labels = x, autopct = "%0.2f%%")
plt.show()
```
